#Â Top 100 hot songs in Spain

```python
# Importing libraries
from bs4 import BeautifulSoup
import pandas as pd
import re
import requests
```

```python
# Defining the URL
url = "https://www.top-charts.com/songs/all-genres/spain/itunes/2021-W47"
```

```python
response = requests.get(url)
```

```python
# Check response status code 
response.status_code
```

```python
#parse and store the contents of the url call
soup = BeautifulSoup(response.content,'html.parser')
```

```python
#prettify the soup 
print(soup.prettify())
```

```python
# Getting the song titles
soup.select("span.song-text div.text-truncate")[0].text
```

```python
# Getting the song artists
soup.select("div.row.artist-text")[0].text
```

```python
# Importing tqdm
from tqdm.notebook import tqdm
```

```python
# defining the columns
title = []
artists = []
```

```python
# checking the length
len_songs = len(soup.select("span.song-text div.text-truncate"))
```

```python
len_songs
```

```python
#creating the dataframe
for i in tqdm(range(len_songs)):
    title.append(soup.select("span.song-text div.text-truncate")[i].text)
    artists.append(soup.select("div.row.artist-text")[i].text)
```

```python
spain_top100 = pd.DataFrame({'title':title, "artists":artists})
```

```python
# checking the sape
spain_top100.shape
```

```python
spain_top100.head()
```
